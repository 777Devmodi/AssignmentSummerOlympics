# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
listofcities = data['City'].unique()
Cities=list(listofcities)
pd.DataFrame(Cities,columns=['City'])
```

```python
type(listofcities)
```

```python
listofrecord=[]
for City in data['City'].unique():
    #print(data['City'])
    listofrecord.append([City ,len(data[data['City'] == City])])
pd.DataFrame(listofrecord ,columns=['City','Count'])
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
Goldlist=data[data['Medal']=='Gold']
```

```python
Goldlist.head()
```

```python
Goldlist['Sport'].unique()
listofgold=[]
for Sport in Goldlist['Sport'].unique():
    listofgold.append([Sport, len(Goldlist[Goldlist['Sport'] == Sport])])
```

```python
listofgold
```

```python
pd.DataFrame(listofgold,columns = ['Sport','GoldMedal']).sort_values(by='GoldMedal', ascending=False).head().plot(x = 'Sport', y = 'GoldMedal', kind = 'bar', figsize = (15,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
Sportarr=[]
for Sport in data['Sport'].unique():
    Sportarr.append([Sport,len(data[data['Sport']== Sport])])
pd.DataFrame(Sportarr,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (15,10))
```

### 4. Which player has won most number of medals? (Top 5)

```python
playername=[]
for Athlete in data['Athlete'].unique():
    playername.append([Athlete,len(data[data[Athlete]==Athlete])])
pd.DataFrame(playername,columns = ['Athlete','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Athlete', y = 'freq', kind = 'bar', figsize = (15,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
Eventlist=[]
for Event in data['Event'].unique():
    Eventlist.append([Event,len(data[data['Event']== Event])])
pd.DataFrame(Eventlist,columns = ['Event','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Event', y = 'freq', kind = 'bar', figsize = (15,10))
```
